```python
import sys
import os
import re


def buildCreate(targetFile,sourceFile,columnFile,numCol,tableName):
    targetList = []
    sourceList = []
    columnList = []
    lineList = []

    arquivo = open('001_CREATE_VIEW_' + tableName + '.txt','w')
    arquivo.write('CREATE OR REPLACE VIEW {} ('.format(tableName))
    with open(targetFile,mode='r',encoding='UTF-8') as file:
        for line in file.readlines():
            targetList.append(line)
    with open(sourceFile,mode='r',encoding='UTF-8') as file:
        for line in file.readlines():
            sourceList.append(line)
    with open(columnFile,mode='r',encoding='UTF-8') as file:
        for line in file.readlines():
            columnList.append(line)
    for x in range(numCol - 1) :
        lineList.append('{}.{} {},'.format(sourceList[x],columnList[x],targetList[x]))
    lineList.append('{}.{} {}'.format(sourceList[numCol-1],columnList[numCol-1],targetList[numCol-1]))
    for item in lineList:
        arquivo.write(item.replace('\n','') + '\n')
    arquivo.close()
    filename = '001_CREATE_VIEW_' + tableName + '.txt'
    arquivo = open(filename, 'a+') # Abra o arquivo (leitura)
    arquivo.close()
    buildQuery(filename)
    includeTables(filename,"join.txt")
    
def buildQuery(filename):
    linesList = []
    with open(filename,mode='r',encoding='UTF-8') as file:
        for line in file.readlines():
            linesList.append(line)
    with open(filename,mode='w',encoding='UTF-8') as file:
        for item in linesList:
            if item[0] == '.':
                file.write(item.replace('.','NULL'))
            file.write(item)
    linesList = []
    with open(filename,mode='r',encoding='UTF-8') as file:
        for line in file.readlines():
            linesList.append(line)
        for item in linesList:
            if item[0] == '.':
                linesList.remove(item)
    with open(filename,mode='w',encoding='UTF-8') as file:
        for item in linesList:
            if item[0] == '.':
                file.write(item.replace('.','NULL'))
            file.write(item)
            
def includeTables(filename,joinFile):
    joinList = []
    with open(joinFile,mode='r',encoding='UTF-8') as file:
        for line in file.readlines():
            joinList.append(line)
    f = open(filename, "a")
    f.write("FROM ")
    for item in joinList:
        f.write(item)
    f.close()
    f = open(filename,'r')
    for line in f.readlines():
        print(line)
    f.close()
            
            
    
```
